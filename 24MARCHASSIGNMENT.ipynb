Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
predicting the quality of wine.

The wine quality dataset is a collection of chemical and sensory attributes from two different types of Portuguese wine. The dataset includes 11 input variables (features) and one output variable (target), which reflects wine quality on a scale of 0 to 10.

The following are the important characteristics of the wine quality dataset:


1. Fixed acidity: The quantity of fixed acids present in the wine that contribute to its tartness. Greater levels of fixed acidity can provide a sour or acidic flavour to a wine.

2. Volatile acidity: The quantity of volatile acids present in the wine that contribute to its fragrance. Wines with higher amounts of volatile acidity may have a vinegar-like aroma.

3. Citric acid: The quantity of citric acid present in the wine that contributes to its acidity and freshness.

4. The quantity of sugar that remains in the wine after fermentation is referred to as residual sugar. Greater residual sugar levels can give a wine a sweet flavour.

5. Chlorides: The quantity of salt in the wine that might affect its taste and texture.

6. The quantity of sulphur dioxide added to the wine to prevent oxidation and microbiological development is known as free sulphur dioxide.

7. Density: The density of the wine is proportional to its alcohol concentration.

8. The acidity or basicity of the wine is measured on a scale of 0 to 14. Lower pH values imply more acidity, whereas higher pH values indicate less acidity.

9. Sulphates: The quantity of sulphates present in the wine, which can operate as an antibacterial and antioxidant.

10. Alcohol percentage: The proportion of alcohol in the wine that contributes to its body and mouthfeel.

1. Missing data in databases, especially the wine quality dataset, is a widespread issue. There are various methods for dealing with missing data, including:

2. Deletion: We eliminate the rows or columns with missing values using this approach. This procedure is simple, but it can result in data loss, especially if the missing data is important.

3. Mean/median imputation: We replace missing values with the mean or median of the available data for the associated characteristic in this approach. This approach is straightforward and quick to use, however it might provide biassed conclusions and minimise data variance.

4. Regression imputation: We utilise regression models to forecast missing values based on the available data for the associated feature in this approach. 5. Multiple imputation: With this approach, we construct numerous imputed datasets, each with a unique set of imputed values, and then combine the results to get a final dataset. This approach works well with missing data, but it is computationally costly and may not perform well if the missing data is not absent at random.

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

```python
df.dropna(axis=0,inplace= True)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

# Perform PCA
pca = PCA()
pca.fit(df_scaled)
```

```python
import numpy as np
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

As we can see 6 components explain approximately 90 % of the variance in the data. So we can reduce the number of features to 6 and retain most of the information.

```python
wine_pca = pca.transform(df_scaled)[:, :6]
```
