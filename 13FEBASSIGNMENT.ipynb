Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The Arithmetic Error class is the root class for any mistakes that occur during numerical operations. It is a subclass of the Exception class and serves as the parent class for various other arithmetic-related exception classes.

Here are two examples:

* ZeroDivisionError: This exception is raised when attempting to divide a number by zero.

* OverflowError: This exception is raised when a numeric calculation exceeds the maximum representable value.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is used to handle errors that occur when a lookup operation fails. It is the base class for all exception classes that indicate that an index, key, or other lookup value is not found in a sequence or mapping object.

* KeyError: This exception is raised when attempting to access a non-existent key in a dictionary. 
* IndexError: This exception is raised when attempting to access an index that is out of range for a sequence.
